```python
import glob

#path =r'data/luftdaten/5331' # use your path

filenames = glob.glob("*.csv")
count_files = 0
li = []
for filename in filenames:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)


us_census = pd.concat(li, axis=0, ignore_index=True)
```

```python
us_census.head(60)
```

```python
us_census['Income'] = us_census['Income'].replace({'\$':''}, regex = True)
```

```python
us_census['Income'] = pd.to_numeric(us_census['Income'] ,errors='coerce')
```

```python
us_census.head(2)
```

```python
us_census[['Males','Females']] = us_census['GenderPop'].str.split("_",expand=True)
```

```python
us_census['Males'] = us_census['Males'].replace({'M':''}, regex = True)
```

```python
us_census['Females'] = us_census['Females'].replace({'F':''}, regex = True)
```

```python
us_census['Males'] = pd.to_numeric(us_census['Males'], errors='coerce')
us_census['Females'] = pd.to_numeric(us_census['Females'] ,errors='coerce')
```

```python
us_census.info()
```

```python
us_census['Females'] = us_census['Females'].fillna(us_census.TotalPop - us_census.Males)
```

```python
import numpy as np
```

```python
us_census["Females"] = us_census["Females"].fillna( method ='ffill', inplace = True)
```
